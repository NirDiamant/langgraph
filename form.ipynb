document.getElementById('feedbackForm').addEventListener('submit', async function(e) {
    e.preventDefault();
    
    const button = e.target.querySelector('button[type="submit"]');
    const successMessage = document.getElementById('successMessage');
    
    // Get form data
    const tutorialRating = document.querySelector('input[name="tutorial_rating"]:checked')?.value;
    const productInterest = document.querySelector('input[name="product_interest"]:checked')?.value;
    const suggestions = document.querySelector('textarea[name="suggestions"]').value;
    const background = document.querySelector('select[name="background"]').value;
    
    // Validate required fields
    if (!tutorialRating) {
        alert('Please rate the tutorial!');
        return;
    }
    if (!productInterest) {
        alert('Please tell us if you\'ll try the product!');
        return;
    }
    
    // Prepare data for Google Forms with YOUR actual entry IDs
    const formData = new FormData();
    formData.append('entry.1442236884', tutorialRating);  // Tutorial Rating
    formData.append('entry.1525387088', productInterest); // Product Interest
    formData.append('entry.929880981', suggestions);      // Suggestions
    formData.append('entry.81513512', background);        // Background
    
    // Update button
    button.disabled = true;
    button.textContent = '⏳ Submitting...';
    
    try {
        // Submit to YOUR Google Form
        await fetch('https://docs.google.com/forms/d/e/1FAIpQLSeCDpJqqI-cx3wG6M2TkWqF27AD_qqSkpHqT4qPzdjWam2B_A/formResponse', {
            method: 'POST',
            body: formData,
            mode: 'no-cors'
        });
        
        // Show success
        this.style.display = 'none';
        successMessage.style.display = 'block';
        
    } catch (error) {
        console.error('Error:', error);
        button.disabled = false;
        button.textContent = '🚀 Submit Feedback';
        alert('There was an issue. Please try again.');
    }
});
</script>
